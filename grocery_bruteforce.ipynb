```python
import pandas as pd
from itertools import combinations
from tqdm.notebook import tqdm

df = pd.read_csv('./Groceries_dataset.csv')
df
```

```python
itemsets = [gr['itemDescription'].tolist() for _, gr in df.groupby(['Member_number', 'Date'])]
members = [(gr['Member_number'].tolist()[0], gr['Date'].tolist()[0]) for _, gr in df.groupby(['Member_number', 'Date'])]
trans = { 'tid': members, 'it': itemsets }
trans = pd.DataFrame(trans)
trans = trans.sample(10)
trans
```

```python
items = []
its = trans['it'].tolist()
for it in its: items += it
items = list(set(items))

itemsets = []
for length in tqdm(range(1, len(items))):
    combs = [set(comb) for comb in list(combinations(items, length))]
    itemsets += combs

len(itemsets)
```

```python
MIN_SUPPORT = 2

freq_itemsets = []

for itemset in tqdm(itemsets):
    sup = trans['it'].apply(lambda x: itemset < set(x)).to_list()
    sup = sum(sup)
    if sup >= MIN_SUPPORT: freq_itemsets.append((itemset, sup))

freq_itemsets
```
